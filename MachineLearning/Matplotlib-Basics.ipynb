```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
import matplotlib as mal
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x,siny)
```

```python
plt.plot(x,cosy)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red")
plt.show()
```

关于color参数：https://matplotlib.org/2.0.2/api/colors_api.html

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

关于linestyle参数：https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0, 1)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
plt.show()
```
